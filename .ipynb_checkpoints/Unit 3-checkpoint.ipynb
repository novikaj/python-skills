### Атрибуты datetime: 
Для начала вспомним, что мы можем делать с datetime. Вот примеры атрибутов, по которым мы можем обращаться к данным объектам:

<li>year: возвращает год</li>
<li>month: возвращает месяц</li>
<li>day: возвращает день</li>
<li>hour, minute, second - час, минута, секунда</li>
<li>dayofweek - день недели, от 0 до 6, где 0 - понедельник, 6 - воскресенье</li>

```python
log.time.dropna()
log['timeday']=0
def daytime(column, data=log):
    for i in range(len(data[column])): 
        hour=data[column][i].hour
        if hour>=0 and hour<6:
            data['timeday'][i]='ночь'
        elif hour>=6 and hour<12: 
            data['timeday'][i]='утро'
        elif hour>=12 and hour<18:
            data['timeday'][i]='день'
        elif hour>=18 and hour<24:
            data['timeday'][i]='вечер'
    return data.head(30)
```

```python
import pandas as pd
log = pd.read_csv('log.csv', sep=';')
log.columns=['user_id', 'time', 'bet', 'win']
log['time']=log['time'].apply(lambda x: str(x).replace('[',''))
log['time']=pd.to_datetime(log['time'])
log['hour']=log['time'].dt.hour
```

```python
import pandas as pd
def fillna_win(row):
    if pd.isna(row.win) and pd.isna(row.bet):
        row.win=0
    elif pd.isna(row.win) and pd.notna(row.bet):
        row.win= -int(row.bet)
    return row
            
        
```

```python
new_win = log.apply(lambda row: fillna_win(row), axis=1)
new_win
```

```python
log['win']=new_win['win']
```

```python
log[log['win']<0].count()
```

```python
log['bet']=log['bet'].fillna(0)
```

```python
log
```

```python
def count_net(row): 
    if row.win<0: 
        return row.win
    else: 
        return row.win-row.bet
```

```python
log_net = log.apply(lambda row: count_net(row), axis=1)
```

```python
log['net'] = log_net
```

```python
log
```

```python
log[log['net']>0].net.median()
```
